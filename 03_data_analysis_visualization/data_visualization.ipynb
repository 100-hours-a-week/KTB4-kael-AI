### 3번 퀘스트
👉 샘플 데이터에서 가입 연도가 2019년 이전인 사람들을 찾아 연봉을 10% 인상한 후, 전체 평균 연봉을 계산하는 코드를 작성하세요.

```python
import pandas as pd

# 샘플 데이터 생성
data = {
    '이름': ['김철수', '이영희', '박민수', '최지현', '홍길동', '정지훈', '이지은'],
    '나이': [25, 30, 35, 28, 40, 50, 22],
    '직업': ['개발자', '마케터', '개발자', '디자이너', 'CEO', '디자이너', '마케터'],
    '연봉': [4000, 3500, 5000, 4200, 10000, 4600, 3300],
    '가입일': ['2020-05-21', '2019-07-15', '2021-01-10', '2018-11-03', '2017-09-27', '2016-04-11', '2022-03-19']
}

df = pd.DataFrame(data)
df['가입일'] = pd.to_datetime(df['가입일'])

mask = df['가입일'].dt.year < 2019
df.loc[mask, '연봉'] = df.loc[mask, '연봉'] * 1.1

print(df)
print(f'\n전체 평균 연봉: {df["연봉"].mean():.1f}')
```

### 3번 퀘스트
👉 평균이 **서로 다른 두 개의 그룹(Group X, Group Y)** 을 비교하는 박스 플롯을 그리세요.

단, **이상값을 강조하고, 스타일을 커스터마이징**해야 합니다.

```python
import matplotlib.pyplot as plt
import numpy as np

# 랜덤 데이터 생성 (두 그룹의 평균 다르게 설정)
np.random.seed(42)
group_x = np.random.randn(50) * 2       # 표준편차 2, 평균 0
group_y = np.random.randn(50) * 2 + 5   # 표준편차 2, 평균 5

flierprops = dict(marker='o', markerfacecolor='red', markersize=8)
medianprops = dict(color='orange', linewidth=2)

plt.boxplot(
    [group_x, group_y],
    labels=['Group X', 'Group Y'],
    flierprops=flierprops,
    medianprops=medianprops,
    patch_artist=True,
    boxprops=dict(facecolor='lightblue')
)
plt.title('커스터마이징 박스 플롯')
plt.ylabel('값')
plt.show()
```

### 1번 퀘스트
👉 `plt.subplots()`를 사용하여 **2x1 형태의 서브플롯을 만들고**, 첫 번째 서브플롯에는 `y = x²`, 두 번째 서브플롯에는 `y = x³`을 그리는 코드를 작성하세요.

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(-5, 5, 100)
y1 = x ** 2  # x의 제곱
y2 = x ** 3  # x의 세제곱

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 8))

ax1.plot(x, y1)
ax1.set_title('y = x²')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x, y2)
ax2.set_title('y = x³')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

plt.tight_layout()
plt.show()
```

### 3번 퀘스트
👉 `gridspec`을 사용하여 **불규칙한 레이아웃의 서브플롯을 생성하고**, 각각 **선 그래프, 산점도, 막대 그래프, 히스토그램**을 그리세요.

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

# 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.random.randn(100)
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 5, 2, 8]

fig = plt.figure(figsize=(12, 8))
gs = gridspec.GridSpec(3, 2)

ax1 = fig.add_subplot(gs[0, :])   # 첫 행 전체: 선 그래프
ax2 = fig.add_subplot(gs[1, 0])   # 두 번째 행 왼쪽: 산점도
ax3 = fig.add_subplot(gs[1:, 1])  # 두 번째~세 번째 행 오른쪽: 막대 그래프
ax4 = fig.add_subplot(gs[2, 0])   # 세 번째 행 왼쪽: 히스토그램

ax1.plot(x, y1)
ax1.set_title('선 그래프')

ax2.scatter(x, y2, s=10, alpha=0.5)
ax2.set_title('산점도')

ax3.bar(categories, values)
ax3.set_title('막대 그래프')

ax4.hist(y2, bins=20)
ax4.set_title('히스토그램')

plt.tight_layout()
plt.show()
```

```python
# 세 개의 과일 집합 정의
set_A = {"사과", "바나나", "체리", "망고"}
set_B = {"바나나", "망고", "포도", "수박"}
set_C = {"망고", "수박", "딸기", "오렌지"}

only_A = set_A - set_B - set_C
only_B = set_B - set_A - set_C
only_C = set_C - set_A - set_B
ab_only = (set_A & set_B) - set_C
ac_only = (set_A & set_C) - set_B
bc_only = (set_B & set_C) - set_A
abc = set_A & set_B & set_C

print(f"A 단독: {len(only_A)}개 → {only_A}")
print(f"B 단독: {len(only_B)}개 → {only_B}")
print(f"C 단독: {len(only_C)}개 → {only_C}")
print(f"A∩B (C 제외): {len(ab_only)}개 → {ab_only}")
print(f"A∩C (B 제외): {len(ac_only)}개 → {ac_only}")
print(f"B∩C (A 제외): {len(bc_only)}개 → {bc_only}")
print(f"A∩B∩C: {len(abc)}개 → {abc}")
```

### 3번 퀘스트
👉 **1번 퀘스트에서 생성한 시계열 데이터에서 이상치를 탐지하고, 이상치만 강조하여 그래프에 표시하는 코드**를 작성하세요.

이상치는 사분위수 범위(IQR)를 이용해 판단합니다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 시계열 데이터 생성
np.random.seed(42)
date_range = pd.date_range(start="2023-01-01", periods=100, freq="D")
values = np.cumsum(np.random.randn(100))

df = pd.Series(values, index=date_range)

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df < Q1 - 1.5 * IQR) | (df > Q3 + 1.5 * IQR)]

plt.plot(df, label='데이터', alpha=0.7)
plt.scatter(outliers.index, outliers.values, color='red', zorder=5, label='이상치')
plt.title('시계열 데이터 이상치 탐지 (IQR)')
plt.xlabel('날짜')
plt.ylabel('값')
plt.legend()
plt.show()
```

👉 평균 `60`, 표준 편차 `15`를 갖는 정규 분포에서 500개의 데이터를 생성한 후, 데이터의 **기본 통계 정보(평균, 표준 편차, 최소값, 최대값)를 출력하는 코드**를 작성하세요.

```python
import numpy as np

# 평균 60, 표준 편차 15인 정규 분포에서 500개의 난수 생성
np.random.seed(42)
data = np.random.normal(loc=60, scale=15, size=500)

print(f'평균:    {data.mean():.2f}')
print(f'표준편차: {data.std():.2f}')
print(f'최솟값:  {data.min():.2f}')
print(f'최댓값:  {data.max():.2f}')
```

### 2번 퀘스트
👉 평균 `50`, 표준 편차 `10`을 갖는 정규 분포에서 특정 값 `x=65`의 **확률 밀도 함수(PDF) 값을 계산하고 출력하는 코드**를 작성하세요.

```python
from scipy.stats import norm

pdf_value = norm.pdf(x=65, loc=50, scale=10)
print(f'x=65의 PDF 값: {pdf_value:.6f}')
```

### 3번 퀘스트
👉 평균 `70`, 표준 편차 `8`을 갖는 정규 분포에서 **(1) 특정 값 `x=80` 이하일 확률을 CDF로 계산하고, (2) 상위 5%에 해당하는 점수를 PPF로 계산하여 출력하는 코드**를 작성하세요.

```python
from scipy.stats import norm

# (1) x=80 이하일 확률 (CDF)
cdf_value = norm.cdf(x=80, loc=70, scale=8)
print(f'x=80 이하일 확률 (CDF): {cdf_value:.4f}')

# (2) 상위 5% 점수 (PPF)
ppf_value = norm.ppf(0.95, loc=70, scale=8)
print(f'상위 5% 점수 (PPF): {ppf_value:.2f}')
```

### 기술 통계 미니퀘스트

### 1번 퀘스트
👉 **주어진 데이터에서 평균과 중앙값의 차이를 계산하는 코드**를 작성하세요.

```python
import numpy as np
import pandas as pd

# 샘플 데이터 생성
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)
df = pd.DataFrame(data, columns=['value'])

mean   = df['value'].mean()
median = df['value'].median()
print(f'평균:   {mean:.2f}')
print(f'중앙값: {median:.2f}')
print(f'차이:   {abs(mean - median):.2f}')
```

### 2번 퀘스트
👉 **데이터에서 이상값(Outlier)을 찾아 제거한 후, 원래 데이터와 이상값 제거 후 데이터의 평균을 비교하는 코드**를 작성하세요.

이상값은 **IQR(사분위 범위)를 사용하여 탐지**하세요.

```python
import numpy as np
import pandas as pd

# 샘플 데이터 생성
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)
df = pd.DataFrame(data, columns=['value'])

Q1  = df['value'].quantile(0.25)
Q3  = df['value'].quantile(0.75)
IQR = Q3 - Q1

df_clean = df[(df['value'] >= Q1 - 1.5 * IQR) & (df['value'] <= Q3 + 1.5 * IQR)]

print(f'원래 데이터 평균:      {df["value"].mean():.2f}')
print(f'이상값 제거 후 평균:   {df_clean["value"].mean():.2f}')
print(f'제거된 데이터 수:      {len(df) - len(df_clean)}')
```

### 3번 퀘스트
👉 **데이터의 왜도(Skewness)와 첨도(Kurtosis)를 계산하여 데이터의 분포 특성을 분석하는 코드**를 작성하세요.

```python
import numpy as np
import pandas as pd

# 샘플 데이터 생성
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)
df = pd.DataFrame(data, columns=['value'])

skewness = df['value'].skew()
kurtosis = df['value'].kurt()

print(f'왜도(Skewness): {skewness:.4f}')
print(f'첨도(Kurtosis): {kurtosis:.4f}')
```

### 2번 퀘스트
👉 평균이 각각 55와 60이고, 표준편차가 8인 두 개의 그룹(A, B) 데이터를 생성한 후, **두 그룹의 데이터 분포를 Seaborn을 활용하여 KDE(커널 밀도 함수)와 함께 히스토그램으로 시각화**하세요.

이후, **두 그룹 간 평균 차이가 유의미한지 t-검정을 수행하는 코드**를 작성하세요.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# 데이터 생성
np.random.seed(42)
group_A = np.random.normal(loc=55, scale=8, size=200)
group_B = np.random.normal(loc=60, scale=8, size=200)

sns.histplot(group_A, kde=True, label='Group A', alpha=0.5)
sns.histplot(group_B, kde=True, label='Group B', alpha=0.5)
plt.title('두 그룹 데이터 분포 비교')
plt.xlabel('값')
plt.legend()
plt.show()

t_stat, p_value = stats.ttest_ind(group_A, group_B)
print(f't-통계량: {t_stat:.4f}')
print(f'p-값:     {p_value:.4f}')
print('결론:', '유의미한 차이 있음 (p < 0.05)' if p_value < 0.05 else '유의미한 차이 없음')
```

### 3번 퀘스트
👉 광고 A를 본 500명 중 120명이 클릭하였고, 광고 B를 본 500명 중 150명이 클릭을 한 데이터가 있습니다.

이 데이터를 바탕으로 **카이제곱 검정을 수행하여 광고 A와 B의 클릭률 차이가 유의미한지 분석하고, Seaborn의 barplot을 사용하여 클릭률을 비교하는 그래프를 그리는 코드**를 작성하세요.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# 데이터 생성 (광고 A와 B의 클릭 여부)
observed_data = pd.DataFrame({
    'Ad_A': [120, 380],
    'Ad_B': [150, 350]
}, index=['Click', 'No Click'])

chi2, p_value, dof, _ = chi2_contingency(observed_data)
print(f'카이제곱 통계량: {chi2:.4f}')
print(f'p-값: {p_value:.4f}')
print('결론:', '유의미한 차이 있음 (p < 0.05)' if p_value < 0.05 else '유의미한 차이 없음')

click_rate = pd.DataFrame({
    '광고': ['Ad_A', 'Ad_B'],
    '클릭률': [120/500, 150/500]
})

sns.barplot(data=click_rate, x='광고', y='클릭률')
plt.title('광고별 클릭률 비교')
plt.ylabel('클릭률')
plt.show()
```

## 시계열 데이터 시각화

#### 데이터 다운로드 (yfinance)

```python
# pip install yfinance
import yfinance as yf
import pandas as pd

# Apple 주가 데이터 (최근 1년)
ticker = yf.Ticker('AAPL')
df = ticker.history(period='1y')

print(f'기간: {df.index[0].date()} ~ {df.index[-1].date()}')
print(f'데이터 수: {len(df)}일')
print()
print(df[['Open', 'High', 'Low', 'Close', 'Volume']].head())
```

#### 종가(Close) 시계열 그래프

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 5))
plt.plot(df.index, df['Close'], label='Close', linewidth=1.5)
plt.title('AAPL 종가 (최근 1년)')
plt.xlabel('날짜')
plt.ylabel('가격 (USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

#### 이동평균 비교 (20일 SMA / 60일 SMA)

```python
import matplotlib.pyplot as plt

df['SMA_20'] = df['Close'].rolling(window=20).mean()
df['SMA_60'] = df['Close'].rolling(window=60).mean()

plt.figure(figsize=(14, 5))
plt.plot(df.index, df['Close'],   label='Close',   linewidth=1, alpha=0.7)
plt.plot(df.index, df['SMA_20'], label='SMA 20일', linewidth=1.8)
plt.plot(df.index, df['SMA_60'], label='SMA 60일', linewidth=1.8)
plt.title('AAPL 종가 + 이동평균')
plt.xlabel('날짜')
plt.ylabel('가격 (USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

#### 거래량(Volume) + 종가 서브플롯

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

ax1.plot(df.index, df['Close'], color='steelblue', linewidth=1.5)
ax1.set_title('AAPL 종가 & 거래량')
ax1.set_ylabel('가격 (USD)')

ax2.bar(df.index, df['Volume'], color='gray', alpha=0.6)
ax2.set_ylabel('거래량')
ax2.set_xlabel('날짜')

plt.tight_layout()
plt.show()
```
